```python
#  Configurar path para importar módulos desde src/
import sys
import os

src_path = os.path.abspath(os.path.join(os.getcwd(), "..", "src"))
if src_path not in sys.path:
    sys.path.append(src_path)

#  Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from data.load_data import load_data, validate_columns

#  Configuración visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

#  Cargar datos
df = load_data()

# Validar columnas esperadas
expected_columns = [
    "work_year", "experience_level", "employment_type", "job_title",
    "salary", "salary_currency", "salary_in_usd", "employee_residence",
    "remote_ratio", "company_location", "company_size"
]
validate_columns(df, expected_columns)

# Vista general
print(f"Shape: {df.shape}")
print("Columnas:", df.columns.tolist())
df.head()
```

```python
# Distribución de salario en USD
sns.histplot(df["salary_in_usd"], bins=30, kde=True, color="skyblue")
plt.title("Distribución de salario en USD")
plt.xlabel("Salario (USD)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Conteo por nivel de experiencia
sns.countplot(x="experience_level", data=df, palette="Set2")
plt.title("Distribución por nivel de experiencia")
plt.xlabel("Nivel de experiencia")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Conteo por tipo de empleo
sns.countplot(x="employment_type", data=df, palette="Set3")
plt.title("Distribución por tipo de empleo")
plt.xlabel("Tipo de empleo")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Top 10 cargos más comunes
top_jobs = df["job_title"].value_counts().head(10)
sns.barplot(y=top_jobs.index, x=top_jobs.values, palette="viridis")
plt.title("Top 10 cargos más comunes")
plt.xlabel("Cantidad")
plt.ylabel("Cargo")
plt.show()
```

```python
# Distribución geográfica de residencia
top_locations = df["employee_residence"].value_counts().head(10)
sns.barplot(x=top_locations.index, y=top_locations.values, palette="coolwarm")
plt.title("Top 10 ubicaciones de residencia")
plt.xlabel("Ubicación")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

```python
# Valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())
```

```python
# 📊 Distribución de modalidad de trabajo (remote_ratio)
sns.countplot(x="remote_ratio", data=df, palette="Blues")
plt.title("Distribución por modalidad de trabajo")
plt.xlabel("Porcentaje de trabajo remoto")
plt.ylabel("Cantidad")
plt.show()
```

```python
# 🌍 Top 10 ubicaciones de la empresa
top_company_locations = df["company_location"].value_counts().head(10)
sns.barplot(x=top_company_locations.index, y=top_company_locations.values, palette="Purples")
plt.title("Top 10 ubicaciones de la empresa")
plt.xlabel("Ubicación")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🏢 Distribución por tamaño de empresa
sns.countplot(x="company_size", data=df, order=["S", "M", "L"], palette="Oranges")
plt.title("Distribución por tamaño de empresa")
plt.xlabel("Tamaño de empresa")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Correlaciones numéricas
numeric_cols = df.select_dtypes(include="number")
corr = numeric_cols.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación entre variables numéricas")
plt.show()
```
